### ðŸŽ¯ Objective
To build a machine learning model that can classify SMS messages as **Spam** or **Ham (Not Spam)** using Natural Language Processing (NLP) techniques.

### ðŸ§  Problem Statement
Spam messages are unsolicited messages sent in bulk, often for advertising or fraudulent purposes. Detecting spam automatically helps protect users and improve communication security.



---
## ðŸ“š Libraries Used
- NumPy
- Pandas
- Matplotlib
- Seaborn
- Scikit-learn


```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```


---
## ðŸ“¥ Dataset Information

The dataset used is the **SMS Spam Collection Dataset**, containing labeled SMS messages.
- Label: spam / ham
- Source: Online public repository


```

url = "https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv"
df = pd.read_table(url, sep='\t', header=None, names=['Label', 'Message'])
df.head()
```

## ðŸ” Dataset Shape and Distribution

```

df.shape
```

```

df['Label'].value_counts()
```

```

sns.countplot(x='Label', data=df)
plt.title("Spam vs Ham Distribution")
plt.show()
```


---
## ðŸ”„ Data Preprocessing

- Convert text labels to numerical values
- Split dataset into training and testing sets
- Apply TF-IDF Vectorization


```

df['Label_Num'] = df['Label'].map({'ham': 0, 'spam': 1})

X_train, X_test, y_train, y_test = train_test_split(
    df['Message'], df['Label_Num'], test_size=0.2, random_state=42
)
```

```

vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```


---
## ðŸ¤– Model Building

We use **Multinomial Naive Bayes**, which is well-suited for text classification problems.


```

model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```


---
## ðŸ“Š Model Evaluation


```

y_pred = model.predict(X_test_tfidf)

print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```


---
## ðŸ§ª Sample Predictions


```

sample_messages = [
    "Congratulations! You have won a free gift",
    "Hey, are we meeting tomorrow?"
]

sample_vec = vectorizer.transform(sample_messages)
predictions = model.predict(sample_vec)

for msg, pred in zip(sample_messages, predictions):
    print(msg, "=>", "Spam" if pred == 1 else "Ham")
```
